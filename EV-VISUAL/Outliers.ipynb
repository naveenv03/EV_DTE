```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb

import os
from IPython.core.display import display
```

```python
df = pd.read_csv("stage0.csv");
display(df.head(5))
```

```python
##  UNIVARIATE ANALYSIS 
```

```python
# city
plt.figure(figsize=(20, 10))
plt.suptitle("Countplots")
plt.subplot(2, 3, 1)
plt.title("City counts")
sb.countplot(df['city'])
#Motor_way
plt.subplot(2, 3, 2)
plt.title("Motor way counts")
sb.countplot(df['motor_way'])

#Country_way
plt.subplot(2, 3, 3)
plt.title("country_roads counts")
sb.countplot(df['country_roads'])

#Drive_Style
plt.subplot(2, 3, 4)
plt.title("driving_style counts")
sb.countplot(df['driving_style'])

#A/C
plt.subplot(2, 3, 5)
plt.title("A/C counts")
sb.countplot(df['A/C'])

#Park_Heater
plt.subplot(2, 3, 6)
plt.title("park_heating counts")
sb.countplot(df['park_heating'])

plt.show()
```

```python
##Observations
##EV's trials are mostly in City and one way, moderately drived.
##EV's trials with a/c is on is very less. so the data is biased towards non-ac trials and with no park heating.
```

```python
# quantity(kWh)
plt.figure(figsize=(20, 10))
plt.suptitle("quantity(kWh)")
plt.subplot(1, 2, 1)
plt.title("Distribution plot")
sb.distplot(df['quantity(kWh)'], kde = True)

plt.subplot(1, 2, 2)
plt.title("Box plot")
sb.boxplot(y=df['quantity(kWh)'])

plt.show()
```

```python
##Abnormaality(outliers) for quant >=40
```

```python
# consumption(kWh/100km)

plt.figure(figsize=(20, 10))
plt.suptitle("consumption(kWh/100km)")
plt.subplot(1, 2, 1)
plt.title("Distribution plot")
sb.distplot(df['consumption(kWh/100km)'], kde = True)

plt.subplot(1, 2, 2)
plt.title("Box plot")
sb.boxplot(y=df['consumption(kWh/100km)'])

plt.show()
```

```python
##observation
##outliers for consumption >30
```

```python
##ecr_deviation: difference between the energy consumption rate recorded at each trip and the value announced by the manufacturer, which is stated as 16.8 kilowatts per hundred kilometers.
```

```python
# ecr_deviation
plt.figure(figsize=(20, 10))
plt.suptitle("ecr_deviation")
plt.subplot(1, 2, 1)
plt.title("Distribution plot")
sb.distplot(df['ecr_deviation'], kde = True)

plt.subplot(1, 2, 2)
plt.title("Box plot")
sb.boxplot(y=df['ecr_deviation'])

plt.show()
```

```python
df['ecr_deviation'].describe()
```

```python
# avg_speed(km/h)

plt.figure(figsize=(20, 10))
plt.suptitle("avg_speed(km/h)")
plt.subplot(1, 2, 1)
plt.title("Distribution plot")
sb.distplot(df['avg_speed(km/h)'], kde = True)

plt.subplot(1, 2, 2)
plt.title("Box plot")
sb.boxplot(y=df['avg_speed(km/h)'])

plt.show()
```

```python
##outliers below 10 kmph and above 80 kmph
```

```python
## outliers
outliers_df = df[
    (df['quantity(kWh)'] > 40) |
    (df['consumption(kWh/100km)'] > 50) |
    (df['avg_speed(km/h)'] < 10) |
    (df['avg_speed(km/h)'] > 80)
]
```

```python
outliers_df.shape
```

```python
outliers_df
```

```python
non_outliers_df = df.drop(index = outliers_df.index)
```

```python
non_outliers_df.to_csv("eda_outliers_removed.csv", index=False)
```
